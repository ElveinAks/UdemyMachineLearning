## Robust Regression

outlier demo http://digitalfirst.bfwpub.com/stats_applet/stats_applet_5_correg.html

Basically, Linear Regression is sensitive to outliers (residuals).

## RANdom SAmple Consensus Algorithm(RANSAC)

http://scikit-learn.org/stable/modules/linear_model.html#ransac-regression



```python
X = df['RM'].values.reshape(-1, 1)
y = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_           ##Selects all the inliers: inlier=True, outlier=False
outlier_mask = np.logical_not(inlier_mask)  ##Turns all the Trues in Falses and all Falses into Trues
                                            ##in array inlier_mask
```

```python
line_x = np.arange(3, 10, 1)                        ##[3,4,5,6,7,8,9]
line_y_ransac = ransac.predict(line_x.reshape(-1, 1))
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,10))
plt.scatter(X[inlier_mask], y[inlier_mask],
           c='blue', marker='o', label='inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
           c='brown', marker='s', label='outliers')
plt.plot(line_x, line_y_ransac, c='green')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper left');
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

## Performance Evaluation of Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values
```

```python
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

## Method One: Residual Analysis

```python
plt.figure(figsize=(12,10))
plt.scatter(y_train_pred, y_train_pred-y_train, c='blue', marker='o', label='Training Data')
plt.scatter(y_test_pred, y_test_pred-y_test, c='red', marker='*', label='Test Data')
plt.xlabel('Predicted Values')
plt.ylabel("Residual Values")
plt.xlim(-10,50)
plt.hlines(y=0, xmin=-10, xmax=50, lw=2)
plt.legend(loc='upper left');
```

## Method Two: Mean-Square Error

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

## Method Three: Coefficient of Determination, $R^{2}$

$R^{2} = 1 - \frac{SSE}{SST}$

<ul>
    <ol>SSE: Sum of Squared Errors</ol>
    <ol>SST: Sum of Squared Totals</ol>
</ul>

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

## What Does A Near-Perfect Model Look Like?

```python
generate_random = np.random.RandomState(0)
x = 10 * np.random.rand(1000)
y = 3 * x + np.random.rand(1000)
plt.figure(figsize=(10,8))
plt.scatter(x,y);
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=0)
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train.reshape(-1,1), y_train)
```

```python
y_train_pred = model.predict(X_train.reshape(-1,1))
y_test_pred = model.predict(X_test.reshape(-1,1))
```

## Method One: Residual Analysis

```python
plt.figure(figsize=(12,10))
plt.scatter(y_train_pred, y_train_pred-y_train, marker='*', c='orange', label='Training Data')
plt.scatter(y_test_pred, y_test_pred-y_test, marker='o', c='blue', label='Test Data')
plt.xlabel('Predicted Data')
plt.ylabel('Residual Data')
plt.hlines(y=0, xmin=-5, xmax=35, lw=2)
plt.xlim([-5, 35])
plt.ylim([-30, 20])
plt.legend(loc='upper left');
```

The y-limit was set to the same range as the scatter plot of the dataset, to show the lessening of the variation a near-perfect correlation has.

## Method Two: Mean-Squared Error

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

## Method Three: Coefficient of Determination, $R^{2}$

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
